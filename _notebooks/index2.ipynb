#### Loading and cleaning the data

```python
# takes 1-3 mins
country_dataframes = {}
metadata_dataframes = {}
country_code_region_df = None
with tarfile.open(r'data\wid_all_data.tar.gz', 'r:gz') as tar:
    for member in tar.getmembers():
        if member.isfile():
            file_obj = tar.extractfile(member)
            if file_obj is None:
                continue
            filename = member.name
            if filename.endswith('.csv'):
                df = pd.read_csv(file_obj,delimiter=';')
                if 'WID_countries' in filename:
                    country_code_region_df = df
                elif 'metadata' in filename:
                    metadata_dataframes[filename.split('/')[1][:-4]] = df
                else:
                    country_dataframes[filename.split('/')[1][:-4]] = df
```

```python
def get_countryname_from_code(code):
    return country_code_region_df[country_code_region_df['alpha2'] == code]['titlename'].to_numpy()[0]

conv_rate = requests.get('https://latest.currency-api.pages.dev/v1/currencies/eur.json').json()

def convert_to_eur(value,code):
    return value/conv_rate['eur'][code.lower()]
```

```python
data = []
for k,v in country_dataframes.items():
    v2 = v.copy()
    gdp_over_years = v2[v2['variable'] == 'agdproi992'][['year','value']]
    md = metadata_dataframes[f'WID_metadata_{k.split('_')[-1]}']
    try:
        country_currency = md[md['variable'] == 'agdproi992']['unit'].to_numpy()[0].lower()
    except:
        continue
    try:
        gdp_over_years['value'] = convert_to_eur(gdp_over_years['value'],country_currency)
    except KeyError:
        continue
    # [Net personal wealth]=[Personal non-financial assets]+[Personal financial assets]-[Personal debt]
    wealth_ineq_gini = v2[(v2['variable'] == 'ghwealj992') & (v2['percentile'] == 'p0p100')][['year','value']]
    population = v2[(v2['variable'] == 'npopuli992') & (v2['percentile'] == 'p0p100')][['year','value']]
    merged_df = reduce(lambda left, right: pd.merge(left, right, on='year',how='inner'), [gdp_over_years, wealth_ineq_gini, population])
    merged_df.sort_values('year',inplace=True)
    if merged_df.empty:continue
    data.append([k,merged_df])
```

Hypothesis: Environmental spending complements growth

Test meanings

| **Metric**      | **Meaning**                               | **Interpretation Thresholds**                |
|-----------------|-------------------------------------------|----------------------------------------------|
| **Correlation** | Strength/direction of linear relationship | -1 (strong negative) to +1 (strong positive) |
| **p-value**     | Statistical significance                  | <0.05 = Significant                          |
| **Coefficient** | Change in GDP growth per 1% spending      | Magnitude shows economic impact              |
| **R-squared**   | Explanatory power of model                | 0-1 (higher = better explanation)            |


```python
def analyze_environment_growth(countries=['IN', 'US', 'CN', 'DE']):
    combined = []
    for country in countries:
        df = country_dataframes[f'WID_data_{country}']
        df = df[df['variable'].isin(['menvgei999', 'mgdproi999'])] 
        df['country'] = country
        combined.append(df)
        
    
    if not combined:
        return pd.DataFrame()
    
    data = pd.concat(combined)
    

    pivot_data = data.pivot_table(index=['country', 'year'], 
                                columns='variable', 
                                values='value').reset_index()
    

    pivot_data['gdp_growth'] = pivot_data.groupby('country')['mgdproi999'].pct_change() * 100
    

    pivot_data['env_spending_pct'] = (pivot_data['menvgei999'] / pivot_data['mgdproi999']) * 100
    

    clean_data = pivot_data.dropna(subset=['gdp_growth', 'env_spending_pct'])

    fig = px.scatter(clean_data, x='env_spending_pct', y='gdp_growth',
                     color='country', trendline='ols',
                     title="Environmental Spending vs GDP Growth",
                     labels={'env_spending_pct': 'Environmental Spending (% of GDP)',
                             'gdp_growth': 'GDP Growth Rate (%)'})
    fig.show()

    results = []
    for country in countries:
        country_data = clean_data[clean_data['country'] == country]
        if len(country_data) > 5:
            corr, p_value = pearsonr(country_data['env_spending_pct'], 
                                   country_data['gdp_growth'])
            
            X = sm.add_constant(country_data['env_spending_pct'])
            model = sm.OLS(country_data['gdp_growth'], X).fit()
            
            results.append({
                'country': country,
                'correlation': corr,
                'p_value': p_value,
                'coef': model.params[1],
                'rsquared': model.rsquared
            })

    results_df = pd.DataFrame(results)
    print("\nStatistical Results:")
    print(results_df)
    
    return clean_data

ss = analyze_environment_growth(['IN','US','CN','DE','ZA','PE','TH'])
```
